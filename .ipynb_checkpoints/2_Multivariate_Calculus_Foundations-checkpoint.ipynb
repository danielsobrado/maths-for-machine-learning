## 1. Understanding Gradients

The gradient points in the direction of steepest ascent and is fundamental to optimization.

```python
# Define a simple 2D function: f(x,y) = x² + y²
def f(x, y):
    return x**2 + y**2

# Analytical gradient: ∇f = [2x, 2y]
def gradient_f(x, y):
    return np.array([2*x, 2*y])

# Create a grid for visualization
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Plot the function and gradient vectors
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# 3D surface plot
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X, Y, Z, alpha=0.8, cmap='viridis')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('f(x,y)')
ax1.set_title('Function: f(x,y) = x² + y²')

# Gradient field
ax2 = fig.add_subplot(122)
contour = ax2.contour(X, Y, Z, levels=15, alpha=0.6)
ax2.clabel(contour, inline=True, fontsize=8)

# Sample gradient vectors
x_sample = np.arange(-2, 3, 0.5)
y_sample = np.arange(-2, 3, 0.5)
for xi in x_sample:
    for yi in y_sample:
        grad = gradient_f(xi, yi)
        # Normalize for better visualization
        grad_norm = grad / (np.linalg.norm(grad) + 1e-8) * 0.3
        ax2.arrow(xi, yi, grad_norm[0], grad_norm[1], 
                 head_width=0.1, head_length=0.1, fc='red', ec='red')

ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('Gradient Field')
ax2.grid(True)
ax2.set_aspect('equal')

plt.tight_layout()
plt.show()
```
